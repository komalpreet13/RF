# Import Required Libraries

```python
# importing the required libraries
from bs4 import BeautifulSoup
import requests
```

# send get request to the webpage server to get the source code of the page

```python
page = requests.get('https://www.dineout.co.in/delhi-restaurants')
```

```python
page
```

# page content

```python
soup = BeautifulSoup(page.content)
soup
```

# Scraping First Name

```python
# First, we will use html tag where we have the first title of the dint-outs.


first_title = soup.find('div',class_="restnt-info cursor")
first_title
```

```python
first_title.text
```

# Scraping First Location

```python
loc = soup.find('div',class_="restnt-loc ellipsis")
loc
```

```python
loc.text
```

# Scraping First Price

```python
sta = soup.find('span',class_="double-line-ellipsis")
sta.text.replace('₹','')
```

```python
sta.text
```

# Scraping Multiple Titles

```python
# Now we have all the tags in which there are the job titles.

# Now we will extract the text from these tags one by one by looping over these tags

titles = [] # emplty list for store the

for i in soup.find_all('div',class_="restnt-info cursor"):
    titles.append(i.text)
    
titles
```

# Scraping Multiple Location

```python
location = [] # emplty list

for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    location.append(i.text)
    
location
```

# Scraping the Multiple Price

```python
price = [] # emplty list

for i in soup.find_all('span',class_="double-line-ellipsis"):
    price.append(i.text.replace('₹',''))
    
price
```

# Scraping the Multiple Images

```python
images = [] 
for i in soup.find_all("img", class_="no-img"):
    images.append(i['data-src'])
    
images
```

```python
# printing length
print(len(titles),len(location),len(price),len(images))
```

```python
# Making dataframe
import pandas as pd
df = pd.DataFrame({'Titles':titles,'Location':location,'Price':price,'Images_url':images})
df
```
